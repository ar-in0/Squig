```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots display inline in the notebook
%matplotlib inline 

df = pd.read_csv('squigV0_e2e_latencies.csv', skipinitialspace=True)

# us to ms
df['Latency_ms'] = df['TimeDelta'] / 1000.0

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df, 
    x='FrameIdx', 
    y='Latency_ms', 
    hue='TestType', 
    s=20, 
    alpha=0.6, 
    linewidth=0
)

# Formatting
plt.title('End-to-End Latency vs Frame Index')
plt.xlabel('Frame Index')
plt.ylabel('Latency (ms)')
plt.grid(alpha=0.3)
plt.legend(title='Test Scenario')

# ignore     startup spikes
plt.ylim(0, 500)

plt.show()
```
